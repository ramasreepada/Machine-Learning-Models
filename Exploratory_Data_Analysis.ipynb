```python
'''This program explains various functions 
applied on haberman dataset'''

import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# load the dataset
haberman = pd.read_csv("haberman.csv")
```

```python
# prints the shape of the dataset
haberman.shape

haberman.head()
```

```python
# print the column names of the dataset
print(haberman.columns) 
```

It is understood that this dataset does not have a heading. The first instance is treated as header. Therefore, the dataset should be read without header

```python
cols = ["age","year","nodes","status"]
haberman = pd.read_csv("haberman.csv",header=None,names=cols)
# to know the dimensions of the haberman dataset
haberman.shape
```

```python
print(haberman.columns)

print(haberman.head())
```

```python
type(haberman)
```

It is understood that the 4th column stores the class label of each instance. In the following line of code, we find the total number of instances for each class. 

```python
haberman["status"].value_counts()
```

From the above result, it is observed that there are two classes and the first class has 225 instances and the second class has 81 instances. Therefore, the dataset is imbalanced.

```python
# data with class label "1"
haberman.loc[haberman["status"] == 1]
```

```python
haberman.loc[haberman["status"] == 2]
```

```python
# to select the instances of each class separately
class_1_instances = haberman.loc[haberman["status"] == 1]
class_2_instances = haberman.loc[haberman["status"] == 2]
```

```python
# pdf of first feature
sb.FacetGrid(haberman,hue="status",height=5).map(sb.distplot,"age") \
.add_legend()
plt.show()
```

```python
# pdf of second feature
sb.FacetGrid(haberman,hue="status",height=5).map(sb.distplot,"year") \
.add_legend()
plt.show()
```

```python
# pdf of third feature
sb.FacetGrid(haberman,hue="status",height=5).map(sb.distplot,"nodes") \
.add_legend()
plt.show()
```

From the pdf of first and second feactures, it can be observed that its not possible to deduce any conclusion from the these features. However, the distribution from the instances of 3rd feature are quite different. 

```python
# pdf and cdf for the first feature values are plotted below.

counts, bin_edges = np.histogram(haberman["age"], bins=10, 

                                 density = True)

print(counts)
pdf = counts/(sum(counts))
# print(pdf);
print(bin_edges);

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf);



plt.show();
```

```python
# pdf and cdf on second feature
counts, bin_edges = np.histogram(haberman["year"], bins=10, 

                                 density = True)

print(counts)
pdf = counts/(sum(counts))
# print(pdf);
print(bin_edges);

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf);

plt.show();
```

```python
# pdf and cdf on third feature
counts, bin_edges = np.histogram(haberman["nodes"], bins=10, 

                                 density = True)

print(counts)
pdf = counts/(sum(counts))
# print(pdf);
print(bin_edges);

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf);

plt.show();
```

```python
# printing mean of each feature w.r.t. each class
print("mean of class 1 istances for each feature")
print(np.mean(class_1_instances.iloc[:,0:3]))

print("mean of class 2 istances for each feature")
print(np.mean(class_2_instances.iloc[:,0:3]))
```

```python
# printing standard deviation of each feature 
# w.r.t. each class

print("standard deviation of class 1 istances")
print(np.std(class_1_instances.iloc[:,0:3]))

print("standard deviation of class 2 instances")
print(np.std(class_2_instances.iloc[:,0:3]))


# printing variance of each feature 
# w.r.t. each class

print("variance of class 1 istances")
print(np.var(class_1_instances.iloc[:,0:3]))

print("variance of class 2 instances")
print(np.var(class_2_instances.iloc[:,0:3]))

```

# Observations and Conclusion:
Using, the feautures age and year it is difficult to identify the class of the instance.
The samples cannot be linearly separated using a these pair of features.
Feature "nodes" is very significant to identify the class (from the probabilistic density function)

However, using feature 2 solely can lead to few misclassifications.
